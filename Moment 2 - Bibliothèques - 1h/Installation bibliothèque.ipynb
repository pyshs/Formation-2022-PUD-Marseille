Vérifier la présence de trios mots (macron, python, covid) sur trois pages de journaux (le monde, libération, le figaro) et sauvegarder chaque page dans un fichier

```python
# Définition des variables
liste_mots_cles = ["macron","covid","python"]
liste_urls = ["https://www.lemonde.fr/",
              "https://www.liberation.fr/",
              "https://www.lefigaro.fr/"]

# Boucle sur les urls
for url in liste_urls:
    # récupérer le contenu
    page = requests.get(url)
    contenu = page.text
    print(len(contenu))
    
    # Boucle sur les mots clés
    for motcle in liste_mots_cles:
        # Nombre d'occurrences
        occurences = contenu.count(motcle)
        # Afficher
        print(url,motcle,occurences)
    
    # Ecrire dans un fichier
    nomdefichier = url.replace("https://www.","").replace(".fr","").replace("/","")
    print(nomdefichier)
    with open(nomdefichier,"w",encoding = "utf8") as fichier:
        fichier.write(contenu)
```
